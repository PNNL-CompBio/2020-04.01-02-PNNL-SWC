# Layers

## Change scatterplot to line plot

```python
from plotnine import geom_line
(ggplot(data=gapminder,
       mapping=aes(x='year',
                   y='lifeExp',
                   by='country',
                   color='continent')) 
 + geom_line())
```

## Layer on a `geom_point` to the plot

```python
from plotnine import geom_line
(ggplot(data=gapminder,
       mapping=aes(x='year',
                   y='lifeExp',
                   by='country',
                   color='continent')) 
 + geom_line()
 + geom_point())
```

## Color by continent only with `geom_line` not `geom_point`

```python
from plotnine import geom_line
(ggplot(data=gapminder,
       mapping=aes(x='year',
                   y='lifeExp',
                   by='country',
                   )) 
 + geom_line(aes(color='continent'))
 + geom_point())
```

# Transformations and Statistics

```python
(ggplot(data=gapminder,
        mapping=aes(x = 'gdpPercap',
                    y = 'lifeExp',
                    color = 'continent'))
+ geom_point())
```

## Add transparency and change the scale of the x axis to $\log_{10}$

```python
from plotnine import scale_x_log10
(ggplot(data=gapminder,
        mapping=aes(x = 'gdpPercap',
                    y = 'lifeExp',
                    color = 'continent'))
+ geom_point(aes(alpha='continent'))
+ scale_x_log10())
```

## Add a linear fit to the data using `geom_smooth`
Move country->color aesthetic to `geom_point` instead of the global `ggplot` aesthetic.  Increase the size of the linear fit.

```python
from plotnine import scale_x_log10, geom_smooth
(ggplot(data=gapminder,
        mapping=aes(x = 'gdpPercap',
                    y = 'lifeExp',
                    ))
+ geom_point(color='orange',
            size=4)
+ scale_x_log10()
+ geom_smooth(method='lm',
             size=1.5))
```

# Multi-panel figures

```python
from IPython.display import HTML
az_mask = ((gapminder['country'].str.startswith('A'))  
           | (gapminder['country'].str.startswith('Z')))
HTML(gapminder[az_mask].to_html())
```

```python
gapminder['A-Z mask'] = az_mask
gapminder
```

```python
from plotnine import facet_wrap
az_countries = gapminder[az_mask]
(ggplot(data   =az_countries,
        mapping= aes(x = 'year',
                     y = 'lifeExp',
                     color = 'continent'))
+ geom_line()
+ facet_wrap( 'country' )
)
```

## Flip the x-axis labels $90^\circ$

```python
from plotnine import facet_wrap, theme, element_text

az_countries = gapminder[az_mask]
(ggplot(data   =az_countries,
        mapping= aes(x = 'year',
                     y = 'lifeExp',
                     color = 'continent'))
+ geom_line()
+ facet_wrap( 'country' )
+ theme(axis_text_x=element_text(rotation=90))
)
```

# Exporting the plot

```python
from plotnine import ggsave
import os
if not os.path.exists( 'results'):
    os.mkdir('results')

lifeExp_plot = (ggplot(data   =az_countries,
        mapping= aes(x = 'year',
                     y = 'lifeExp',
                     color = 'continent'))
+ geom_line()
+ facet_wrap( 'country' )
+ theme(axis_text_x=element_text(rotation=90))
)

ggsave(filename='results/lifeExp_by_year.jpg',
      plot = lifeExp_plot,
      width = 12,
      height= 10,
      dpi = 300,
      )
```
